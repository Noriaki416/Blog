```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pre_prob = []
post_prob = []
a = 90
b = 20
x = 0.1
for _ in range(1000):
    pre_prob.append(x)
    post_prob.append((a*x)/(a*x + b*(100-x))*100)
    x += 0.1
plt.plot(pre_prob, post_prob)
plt.xlabel('pre_prob')
plt.ylabel('post_prob')
plt.title('sensitivity:{},  specificity:{}'.format(a, 100-b))
plt.savefig('./fig/sensitivity_specificity.jpg')
plt.show()
```
